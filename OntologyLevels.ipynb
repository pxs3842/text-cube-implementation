```python
### Install RDFLib
import rdflib
```

```python
### Form a graph from the owl file.
g = rdflib.Graph()
g.load("/Users/poojasharma/Desktop/FinalProject/root-ontology.owl")
```

```python
### Read all the subjects, predicates and objects present in the graph.
for s,p,o in g:
    print(s.encode('ascii', 'ignore'),p.encode('ascii', 'ignore'),o.encode('ascii', 'ignore'))
```

```python
### SPARQL queries to get the sub-classes of the main class- Thing.

owlURI = 'http://www.w3.org/2002/07/owl#Thing'

def subClassParser(URI):
    qres = g.query('SELECT * Where { ?a <http://www.w3.org/2000/01/rdf-schema#subClassOf>' + ' <'+ URI+ '> }')
    results = []
    for row in qres:
        results.append(row[0][0:])
    
    return results

def labelParser(URI):
    qres = g.query('SELECT * Where { <'+ URI+ '> <http://www.w3.org/2000/01/rdf-schema#label> ?c }')
    results = []
    for row in qres:
        results.append(row[0][0:])
    
    return results[0] ## Return first row since only one label per URI

for u in subClassParser(owlURI):
    print(labelParser(u))
```

```python
### Get First level of hierarchy from the owl which are dimensions
Dimension_URI = []
Dimension_Labels = []

### First level of Hierarchy
Dimension_URI = subClassParser(owlURI)
for u in Dimension_URI:
    Dimension_Labels.append(labelParser(u))
for i in range(len(Dimension_Labels)):
    print(Dimension_URI[i].encode('ascii', 'ignore'), Dimension_Labels[i].encode('ascii', 'ignore'))
```

```python
Dimension_Lbls = []
for i in range(len(Dimension_Labels)):
    Dimension_Lbls.append(Dimension_Labels[i].encode('ascii', 'ignore'))
print(Dimension_Lbls)
```

```python
### First level of Hierarchy for class 'Apple' - index 0

URI = Dimension_URI[0]
FirstLevel_Apple_Labels = []
FirstLevel_Apple_Lbls = []
FirstLevel_Apple_URI = subClassParser(URI)
for u in FirstLevel_Apple_URI:
    FirstLevel_Apple_Labels.append(labelParser(u))
for i in range(len(FirstLevel_Apple_Labels)):
    FirstLevel_Apple_Lbls.append(FirstLevel_Apple_Labels[i].encode('ascii', 'ignore'))
    print(FirstLevel_Apple_URI[i].encode('ascii', 'ignore'), FirstLevel_Apple_Labels[i].encode('ascii', 'ignore'))
```

```python
### Second level of Hierarchy for all 'Apple' sub classes

SecondLevel_Apple_Lbls = []

def SecondLevel_Apple(FirstLevel_Apple):
    URI = FirstLevel_Apple
    SecondLevel_Apple_URI = []
    SecondLevel_Apple_Labels = []
    SubCategories = []
    
    ### For URI in FirstLevel_Apple_URI:
    SecondLevel_Apple_URI.append(subClassParser(URI))

    import itertools
    SecondLevel_Apple_URI = list(itertools.chain.from_iterable(SecondLevel_Apple_URI))
    len(SecondLevel_Apple_URI)
    
    for u in SecondLevel_Apple_URI:
        SecondLevel_Apple_Labels.append(labelParser(u))
    for i in range(len(SecondLevel_Apple_Labels)):
        SubCategories.append(SecondLevel_Apple_Labels[i].encode('ascii', 'ignore'))
    
    SecondLevel_Apple_Lbls = SubCategories
    return SubCategories

for FirstLevel_Apple in FirstLevel_Apple_URI:
    print("Sub-categories of "+labelParser(FirstLevel_Apple)+ " are:")
    print '\n'.join(SecondLevel_Apple(FirstLevel_Apple))
    SecondLevel_Apple_Lbls+=SecondLevel_Apple(FirstLevel_Apple)

print("---------")
print(SecondLevel_Apple_Lbls)
```

```python
### First level of Hierarchy for class 'Features' - index 1

URI = Dimension_URI[1]
FirstLevel_Features_Labels = []
FirstLevel_Features_Lbls = []
FirstLevel_Features_URI = subClassParser(URI)
for u in FirstLevel_Features_URI:
    FirstLevel_Features_Labels.append(labelParser(u))
for i in range(len(FirstLevel_Features_Labels)):
    FirstLevel_Features_Lbls.append(FirstLevel_Features_Labels[i].encode('ascii', 'ignore'))
    print(FirstLevel_Features_URI[i].encode('ascii', 'ignore'), FirstLevel_Features_Labels[i].encode('ascii', 'ignore'))
```
